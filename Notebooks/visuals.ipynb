```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Visualize data

## Data correction for data sets:

### 16-june

low data names: ['11', '12', '13', '14', '15', '16', '17', '18', '113', '114', '115', '116', '121', '122', '123', '124' ]
<br>
angle corrections for low data: [-3, 3, 2.5, -9, 2, -7, -3, 0, 8, 6, 2, 9.5, -5, -6.5, 5, -1]

```python
fileName = '../TestData/17-june/test2/raw/a.csv'

lidarData = []
with open(fileName, 'r') as file: 
    csvReader = csv.reader(file)
    
    for row in csvReader:
        if (float(row[0]) < 350):
            lidarData.append([float(row[0]),float(row[1])])
        
lidarData = np.array(lidarData)

#get cartesian coordinates
extraTheta = 0
x = -lidarData[:,1]*np.cos((lidarData[:,0]+ extraTheta)*math.pi/180)/1000
y = -lidarData[:,1]*np.sin((lidarData[:,0]+ extraTheta)*math.pi/180)/1000

#plot
plt.scatter(x, y, s=5)
plt.scatter([0],[0], s=10, color='red')
```

# Generate visuals for all data and save it (16 june)

```python
lowFilesNames = ['11', '12', '13', '14', '15', '16', '17', '18', '113', '114', '115', '116', '121', '122', '123', '124' ]
lowFilesCorrections = [-3, 3, 2.5, -9, 2, -7, -3, 0, 8, 6, 2, 9.5, -5, -6.5, 5, -1]

highFilesNames = ['111h', '112h', '113h', '114h', '116h', '11h', '13h', '14h', '19h']

for i,name in enumerate(lowFilesNames):
    # define names of the file with the data and the new generated image
    filePath = f'../TestData/16-june/{name}.csv'
    imagePath = f'../TestData/16-june/visuals/low/{name}.jpg'

    #locad file data
    lidarData = []
    with open(filePath, 'r') as file:
        csvReader = csv.reader(file)

        for row in csvReader:
            lidarData.append([float(row[0]),float(row[1])])
    lidarData = np.array(lidarData)
    
    #get cartesian coordinates
    x = lidarData[:,1]*np.cos((lidarData[:,0]+ lowFilesCorrections[i])*math.pi/180)/1000
    y = lidarData[:,1]*np.sin((lidarData[:,0]+ lowFilesCorrections[i])*math.pi/180)/1000

    #save new image
    plt.clf()
    plt.scatter(x, y, s=5, color='blue')
    plt.scatter([0],[0], s=10, color='red')
    plt.savefig(imagePath)
    
for name in highFilesNames:
    # define names of the file with the data and the new generated image
    filePath = f'../TestData/16-june/{name}.csv'
    imagePath = f'../TestData/16-june/visuals/high/{name}.jpg'

    #locad file data
    lidarData = []
    with open(filePath, 'r') as file:
        csvReader = csv.reader(file)

        for row in csvReader:
            lidarData.append([float(row[0]),float(row[1])])
    lidarData = np.array(lidarData)
    
    #get cartesian coordinates
    x = lidarData[:,1]*np.cos(lidarData[:,0]*math.pi/180)/1000
    y = lidarData[:,1]*np.sin(lidarData[:,0]*math.pi/180)/1000

    #save new image
    plt.clf()
    plt.scatter(x, y, s=5, color='blue')
    plt.scatter([0],[0], s=10, color='red')
    plt.savefig(imagePath)
```
